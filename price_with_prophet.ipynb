```python
import pandas as pd
from prophet import Prophet
import FinanceDataReader as fdr

# 엔화 대비 달러 환율 정보 가져오기
exchange_rate_usd_jpy = fdr.DataReader('USD/JPY', '2000-01-01', '2024-03-23')
exchange_rate_usd_jpy.reset_index(inplace=True)
data = exchange_rate_usd_jpy[['Date','Close']]

# Prophet 모델을 사용하여 예측하는 함수
def predict_forecast(data):
    data.columns = ['ds', 'y']
    model = Prophet()
    model.fit(data)
    future = model.make_future_dataframe(periods=365)[-365:]
    future.reset_index(inplace=True)
    forecast = model.predict(future)

    return forecast, model

train_data = data[-80:]
train_data.reset_index(inplace=True)
train_data = train_data[['Date','Close']]
forecast = predict_forecast(train_data)
```

```python
forecast[0][['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head(5)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

# 예측 결과 시각화
fig = plot_plotly(forecast[1], forecast[0])
fig.update_layout(title='USD/JPY Exchange Rate Prediction')
fig.show()

# 추세, 주간 및 연간 구성 요소 시각화
fig = plot_components_plotly(forecast[1], forecast[0])
fig.show()
```

```python
import pandas as pd
from prophet import Prophet
import FinanceDataReader as fdr

# 엔화 대비 달러 환율 정보 가져오기
tencent_stock = fdr.DataReader('0700.HK', '2010-01-01', '2024-03-23')
tencent_stock.reset_index(inplace=True)
data = tencent_stock[['Date','Close']]

# Prophet 모델을 사용하여 예측하는 함수
def predict_forecast(input_data):
    input_data.columns = ['ds', 'y']
    model = Prophet()
    model.fit(input_data)
    future = model.make_future_dataframe(periods=365*4)[-365*4:]
    future.reset_index(inplace=True)
    forecast = model.predict(future)

    return forecast, model

train_data = data[-365*6:]
train_data.reset_index(inplace=True)
train_data = train_data[['Date','Close']]
forecast = predict_forecast(train_data)
```

```python
forecast[0][['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

# 예측 결과 시각화
fig = plot_plotly(forecast[1], forecast[0])
fig.update_layout(title='Tencent Stock Price Prediction')
fig.show()

# 추세, 주간 및 연간 구성 요소 시각화
fig = plot_components_plotly(forecast[1], forecast[0])
fig.show()
```
